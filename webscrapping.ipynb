```python
# for page 1 through https://muckrack.com/beat/singapore
from bs4 import BeautifulSoup
import requests
Journalist_Names_1 = []
url  = 'https://muckrack.com/beat/singapore'
r = requests.get(url)
html_doc = r.text
soup = BeautifulSoup(html_doc)
for x in soup.findAll('div', {'class':'mr-directory-item'}):
        Journalist_Names_1.append(x.find('a')['href'])
# soup
```

```python
# for page after 1 through https://muckrack.com/beat/singapore using ?page
from bs4 import BeautifulSoup
import requests
Journalist_Names_2 = []
for i in range(2, 29):
    url  = 'https://muckrack.com/beat/singapore?page='+ str(i)
    r = requests.get(url)
    html_doc = r.text
    soup = BeautifulSoup(html_doc)
    for x in soup.findAll('div', {'class':'mr-directory-item'}):
        Journalist_Names_2.append(x.find('a')['href'])
# soup
```

```python
Journalist_Names = Journalist_Names_1+ Journalist_Names_2
Journalist_Names
```

```python
from bs4 import BeautifulSoup
import requests
Last_Date, Name, Status, DigitalReporter, ReporterLocation, Reporter_link_First, Reporter_link, account_9, Beat,Covers = [],[],[],[],[],[],[],[],[],[]
for i, x in enumerate(Journalist_Names):
    print(i,x)
    url  = 'https://muckrack.com'+x
    r = requests.get(url)
    html_doc = r.text
    soup = BeautifulSoup(html_doc)
    N = soup.select('strong:contains("Covers:")')
    if N!=None:
        try:
            Covers.append([i.next_sibling.strip() for i in soup.select('strong:contains("Covers:")')])
        except:
            Covers.append('Not Available')
    x = soup.find("div", {"class":"news-story-meta col-xs-12"})
    if x!=None:
        Last_Date.append(soup.find("div", {"class":"news-story-meta col-xs-12"}).find('a').text.strip())
    else:
        Last_Date.append('Not Available')
    a = soup.find("h1", {"class":"profile-name mr-font-family-2 top-none"})
    if a!=None:
        Name.append(soup.find("h1", {"class":"profile-name mr-font-family-2 top-none"}).text.strip())
    else:
        Name.append('Not Available')
    b = soup.find("small", {"class":"profile-verified text-uppercase color-green"})
    if b!=None:
        Status.append(soup.find("small", {"class":"profile-verified text-uppercase color-green"}).text.strip())
    else:
        Status.append('Not Available')
    c = soup.find("div", {"class":"person-details-item person-details-title"})
    if c is not None:
        try:
            DigitalReporter.append(soup.find("div", {"class":"person-details-item person-details-title"}).find('a').text)
        except:
            DigitalReporter.append('Not Available')
    d = soup.find("div", {"class":"person-details-item person-details-location"})
    if d!=None:
        ReporterLocation.append(soup.find("div", {"class":"person-details-item person-details-location"}).find("div").text.strip())
    else:
        ReporterLocation.append('Not Available')
    e = soup.find("span", {"class":"sharing-counts small"})
    if e!=None:
        try:
            Reporter_link_First.append(soup.find("span", {"class":"sharing-counts small"}).find('a')['data-link'])
        except:
            Reporter_link_First.append('Not Available')
    f = soup.find("span", {"class":"sharing-counts small"})
    if f!=None:
        Reporter_link.append(soup.find("span", {"class":"sharing-counts small"}).find('a')['data-url'])
    else:
        Reporter_link.append('Not Available')
    G = soup.find("div", {"class":"profile-section-social"})
    l = []
    if G!=None:
        for link in G.find_all('a'):
            l.append(link.get('href'))
        if len(l)==0:
            account_9.append(['Not Available'])
        if len(l)!=0:
            account_9.append(l)
    else:
        account_9.append(['Not Available'])
    h = []
    H = soup.find("div", {"class":"person-details-item person-details-beats"})
    if H!=None:
        try:
            for link in H.find_all('a'):
                h.append(link.text)
            if len(h)==0:
                Beat.append('Not Available')
            if len(h)!=0:
                Beat.append(h)
        except:
            Beat.append(['Not Available'])
```
